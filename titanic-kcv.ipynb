## Check null values

```python
print(df_train['Age'].isnull().sum())
print(df_train['Cabin'].isnull().sum())
print(df_train['Embarked'].isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

def clean_dataset(dataset):
    
  imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
  dataset['Age'] = imputer.fit_transform(dataset[['Age']])
  imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
  dataset['Embarked'] = imputer.fit_transform(dataset[['Embarked']])
  dataset['Cabin'] = dataset['Cabin'].fillna('N')
  return dataset
```

```python
df_train = clean_dataset(df_train)
```

```python
df_test = clean_dataset(df_test)
```

```python
df_test.info()
```

## Label Encoder

```python
def le_features(col,unique_val,dataset):
    le = LabelEncoder()
    le.fit(list(dataset[col].unique()))
    dataset[col] = le.transform(dataset[col])
    return dataset
```

```python
categorical_features = ['Sex','Name','Ticket','Cabin','Embarked']
```

```python
for f in categorical_features:
    df_train = le_features(f,df_train[f],df_train)
```
